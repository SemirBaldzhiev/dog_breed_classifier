# Dog Breed Classifier

Download dataset from here [Dog Breeds Dataset](https://www.kaggle.com/datasets/khushikhushikhushi/dog-breed-image-dataset)

The dataset contains collection of images for 10 different dog breeds, including:
- **Golden Retriever**
- **German Shepherd**
- **Labrador Retriever**
- **Bulldog**
- **Beagle**
- **Poodle**
- **Rottweiler**
- **Yorkshire Terrier**
- **Boxer**
- **Dachshund**

The task is to classify image of a dog in one of this breeds above


## Imports

```python
import os
import pandas as pd
import numpy as np
from PIL import Image
```

## Constants

```python
BASE_DATASET_DIR = './dataset'
```


## Load images from the dataset

```python
def load_data():
    images = []
    labels = []

    for breed_folder in os.listdir(BASE_DATASET_DIR):
        breed_path = os.path.join(BASE_DATASET_DIR, breed_folder)
        
        if os.path.isdir(breed_path):
            for img_file in os.listdir(breed_path):
                img_path = os.path.join(breed_path, img_file)
                img = Image.open(img_path)
                img = img.resize((150, 150))
                images.append(np.array(img))
                labels.append(breed_folder)
    
    return np.array(images), np.array(labels)
```

```python
X, y = load_data()

print(X.shape, y.shape)
```
